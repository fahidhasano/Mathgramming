```python
#Example: The probability that a patient recovers from a diseases is 0.9. What is the probability that
#an exactly 5 out of the next seven patients will survive?
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
n, p =7, 0.9
binom.pmf(5, n, p, loc=0)
x = np.arange(binom.ppf(0.01, n, p),
              binom.ppf(0.99, n, p))
ax.plot(x, binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
var
```

```python
#A bulb manufacturing company claims that the average
#longevity of their bulb is 3 years. It is given that the variance of longevity
#of manufactured bulb is 0.5 years. A random sample of 35 bulbs gave a
#mean longevity of 2.6 years. Does the sample mean justify the claim of
#the manufacturer? Test this at 5% level of significance. Also find 95%
#confidence interval for population mean.
from scipy.stats import norm
mu_0=3
var=0.5
n=35
xbar=2.6
alpha=5/100
#It's a two tailed test
calculated_z=(xbar-mu_0)/((var/n)**0.5)
z_1_minus_alpha_by_2=norm.ppf(1-(alpha/2))
z_alpha_by_2=norm.ppf(alpha/2)
if calculated_z >z_1_minus_alpha_by_2 or calculated_z< z_alpha_by_2:
    print("Their claim was wrong")
else:
    print("Their claim was right")

x=norm.ppf(((1-(95/100))/2)+(95/100))
interval=1.96*((var/n)**0.5)
print(f"Confidence interval is : ({-interval+xbar},{interval+xbar})")
```

```python
#The average IQ of university women in Bangladesh is
#suspected to be 110. A random sample of 64 women yielded an average
#IQ of 115.5 and a standard deviation of 20. Can you conclude the average
#score of the women in the population is really more than 110? Test this at
#1% level of significance.
from scipy.stats import norm
mu_0=110
sigma=20
n=64
xbar=115.5
alpha=1/100
#It's a one tailed test
calculated_z=(xbar-mu_0)/(sigma/(n**0.5))
z_1_minus_alpha=norm.ppf(1-(alpha))
if calculated_z >z_1_minus_alpha :
    print("the average score of the women in the population is more than 110")
else:
    print("The average IQ of university women in Bangladesh is 110")
```

```python
#It is stated by a company that the daily average production
#(in terms of number) of the workers is 20. A sample of 10 workers is taken
#from that company and their productions are as: 40, 35, 48, 43, 35, 39, 44,
#37, 42, 38. At 5% level of significance test the statement of the company
#assuming the production follows normal distribution.
import numpy as np
from scipy.stats import t
data = np.array([40,35,48,43,35,39,44,37,42,38])
s=np.std(data,ddof=1)
n=10
mu_0=20
alpha=5/100
x_bar=np.average(data)
test_statistic=(x_bar-mu_0)/(s/(n)**0.5)
# two tailed test
t_value1= t.ppf(1-(alpha/2),n-1, loc=0, scale=1)

if abs(test_statistic)>t_value1 :
    print("Company is wrong")
else:
    print("Company is right")
```
