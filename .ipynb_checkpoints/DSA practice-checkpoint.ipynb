```python
#linear search for minimum element in an array
a=[3,6,7,9,100,0,29,1,10]
def min(x):
    m=x[0]
    for i in x:
        if m>i:
            m=i
    return m
print(min(a))
```

```python
#binary search for searching the position of a specific element in an array

a=[3,6,7,9,100,0,29,1,10]
def pos(x,tar):
    x.sort()
    print("sorted array:",x)
    l=0
    r=len(x)-1
    while l<=r:
        m=(l+r)//2
        if x[m]==tar:
            return print(tar,"is in ",m,"th position of the array")
        elif x[m]>tar:
            r=m-1
        else:
            l=m+1
    return print("Look in the sorted array!",tar,"does not exist in the array")
    
pos(a,3)

```

```python
#Fibonacci using recursion algorithm(2)
# fibonacci(any number) it indexed with 0 so it goes upto any number+1 th
def fibonacci(n):
    if n<=1:
        return n
    else:
        return fibonacci(n-1)+fibonacci(n-2)
fibonacci(50)
```

```python
#euclidean algorithm
def gcd(a,b):    
    r=a%b
    while(r>0):
        a=b
        b=r
        r=a%b
    return b
gcd(50,15)  
    
    
    
```

```python
a = [7, 12, 9, 11, 3, 2]
def sorting_by_selectionsort(x):
    b=[]
    f=x.copy()
    while len(f)>0:
        m=f[0]
        for i in range(len(f)):
            if m>f[i]:
                m=f[i]
        b.append(m)
        f.remove(m)
    return b
sorting_by_selectionsort(a)
```

```python
#stacks
class stacks:
    def __init__(self):
        self.list=[]
    def push(self,elem):
        self.list.append(elem)
    def remove (self):
        if len(self.list)==0:
            return "empty"
        return self.list.pop()
    def peek(self):
        if len(self.list)==0:
            return "empty"
        return self.list[len(self.list)-1]
    def isEmpty(self):
        if len(self.list)!=0:
            return False
        return True
x=stacks()
x.push(5)
x.isEmpty()
x.push(66)
x.list
x.peek()
```

```python
#queue
class queue:
    def __init__(self):
        self.list=[]
    def enqueue(self,elem):
        self.list.append(elem)
    def dequeue(self):
        if len(self.list)==0:
            return "empty"
        return self.list.pop(0)
    def size(self):
        return len(self.list)
    def peek(self):
        return self.list[0]
x=queue()
x.enqueue(5)
x.enqueue(10)
x.enqueue(15)
x.enqueue(20)
print(x.list)
x.dequeue()
x.peek()

        




        
        
```
