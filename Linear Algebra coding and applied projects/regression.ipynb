```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure( figsize=[8,8])
x=3
a=np.arange(-2, 2, 0.2).reshape((-1, 1))
b=a*x+np.random.randn(*a.shape)
plt.plot(a, a*x,'--', label='True')
plt.plot(a, b,'x', label='Noisy')
#U, S, Vh = np.linalg.svd(a, full_matrices=False)
#regr=np.transpose(Vh)@ np.linalg.inv(np.diag(S))@ np.transpose(U)@b
#alternative
regr = np.linalg.pinv(a)@b
plt.plot(a, a*regr,'-.', label='Regression')
plt.xlabel('a')
plt.ylabel('b')
plt.title("Regression Plot")
plt.legend()
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np
a=np.loadtxt(os.path.join("hald_ingredients.csv"),delimiter=',', skiprows=1)
b=np.loadtxt(os.path.join("hald_heat.csv"),delimiter=',' ,skiprows=1)
B =np.linalg.pinv(a)@b
c=a@B
plt.plot( b, label='true')  # etc.
plt.plot( c, label='regression')
plt.xlabel('Sample ingredients')
plt.ylabel('Heat probability')
plt.title("Actual vs Predicted Heat")
plt.legend()
```
