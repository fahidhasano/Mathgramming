```python
#I use hq720.jpg in the same folder as sample:
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
img=np.array(Image.open("hq720.jpg"))
X=np.mean(img, axis=2)
plt.figure(figsize=(20, 20))
plt.imshow(X, cmap='gray')
```

```python
U, S, Vh = np.linalg.svd(X, full_matrices=False)
s=np.diag(S)
j=0
for rank in  [5,100,500]:
    xapproximate=U[:,:rank]@s[:rank,:rank]@Vh[:rank,:]
    plt.figure(j+1)
    plt.figure(figsize=(20, 20))
    plt.title('Rank '+str(rank))
    plt.imshow(xapproximate, cmap='gray')
    j+=1
    
    
```
