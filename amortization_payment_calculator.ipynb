```python
#Answer:
import pandas as pd
def amortization_schedule_payment(principal, nominal_rate, compound, payment_times_each_year, duration):
    r=2
    i=((1+(nominal_rate/compound))**(compound/payment_times_each_year))-1
    op=principal
    n= payment_times_each_year* duration
    R=round((op)/((1-((1+i)**(-n)))/i), r)
    row=[]
    for N in range(1, n+1):
        interest=round(op*i,r)
        pr=round(R-interest,r)
        if n==N:
            R=round(interest+op,r)
            pr=round(R-interest,r)
        op=op-pr   

        row.append([N,R,interest,pr,op])

    df = pd.DataFrame(row, columns=["Payment number","Periodic Payment", "Interest","Principal Repaid","Outstanding Principal"])
    return df               
amortization_schedule_payment(2000, 0.24,12, 4, 2)
```
